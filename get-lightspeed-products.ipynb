```python
import requests
import json
```

```python
BASE_URL = "https://api.webshopapp.com"
API_KEY = "9753631d6b313e68d0fb095f9246b754"
API_SECRET = "4ac91498f0935567a32022759f0a5e9a"
```

```python
def get_data(type):
    page = 1
    all_items = []

    while True:
        print(f"Fetching page {page}...")
        url = f"{BASE_URL}/nl/{type}.json?limit=250&page={page}"
        response = requests.get(url, auth=(API_KEY, API_SECRET))

        if not response.ok:
            print(f"Error: {response.status_code} - {response.text}")

        data = response.json()
        items = data.get(type, [])

        if not items:
            break

        all_items.extend(items)

        print(f"Fetched {len(items)} (of {len(all_items)}) {type} on page {page}.")

        if len(items) < 250:
            break

        page += 1

    print(f"Total {type}: {len(all_items)}")

    return all_items
```

```python
all_products = get_data("products")

with open("lightspeed_products.json", "w") as f:
    json.dump(all_products, f, indent=2)
```

```python
all_variants = get_data("variants")

with open("lightspeed_variants.json", "w") as f:
    json.dump(all_variants, f, indent=2)
```

```python
# Merge
variant_product_id_map = {}

print("Creating product ID to variants map...")

for variant in all_variants:
    product_id = variant["product"]["resource"]["id"]

    if product_id not in variant_product_id_map:
        variant_product_id_map[product_id] = []

    variant_product_id_map[product_id].append(variant)

print("Merging variants into products...")

for product in all_products:
    product_id = product["id"]
    product["variants"] = variant_product_id_map.get(product_id, [])

print("Saving merged products with variants...")

with open("lightspeed_products_with_variants.json", "w") as f:
    json.dump(all_products, f, indent=2)
```
